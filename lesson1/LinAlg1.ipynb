__1.__ Исследовать на линейную зависимость:
$$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$$

__Решение:__

Заметим, что
$$f_{4}(x)=-f_{1}(x)-f_{2}(x)+f_{3}(x)$$

т.к. можно выразить одну функцю через другие, то можно сказать, что они линейно зависимы

__2.__ Исследовать на линейную зависимость:
$$f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}.$$

__Решение:__

Заметим, что
$$f_{4}(x)=(x+1)^2=x^2+2x+1$$
$$f_{4}(x)=f_{3}(x)+2 \cdot f_{2}(x)+0.5 \cdot f_{1}(x)$$

т.к. можно выразить одну функцю через другие, то можно сказать, что они линейно зависимы

__3.__ Найти координаты вектора $x = (2, 3, 5)\in \mathbb{R}^{3}$ в базисе $b_{1}=(0, 0, 10)$, $b_{2}=(2, 0, 0)$, $b_{3}=(0, 1, 0)$.

$$x=(2, 3, 5)=(2,0,0)+(0, 3, 0)+(0, 0, 5)=1\cdot(2, 0, 0)+3\cdot(0, 1, 0)+0.5\cdot(0, 0, 10)=1b_{2}+3b_{3}+0.5b_{1},$$

то есть координатами вектора $x$ в базисе $b_1,b_2,b_3$ являются ($0.5$, $1$, $3$).

__4.__ Найти координаты вектора $3x^{2}-2x+2\in\mathbb{R}^{3}[x]$:

   а) в базисе $1$, $x$, $x^{2}$;

   б) в базисе $x^{2}$, $x-1$, $1$.

__Решение:__

a) в базисе $1$, $x$, $x^{2}$; 
$$(2,-2,3)$$

б) в базисе $x^{2}$, $x-1$, $1$

$$3x^{2}-2x+2 = 3x^2-2(x-1)+0$$
$$(3,-2,0)$$

__5.__ Установить, является ли линейным подпространством:

а) совокупность всех векторов трехмерного пространства, у которых по крайней мере одна из первых двух координат равна нулю;
    
б) все векторы, являющиеся линейными комбинациями данных векторов $\{u_{1}, u_{2}, \ldots, u_{n}\}$.

__Решение:__

a) Проведя проверку по приведенному утверждению, мы увидим, что 

$$(0, a, b)+(c, 0, d) = (0+c, a+0, b+d) = (c, a, b+d)$$ 

Полученные векторы уже не принадлежат указанному в задании множеству всех векторов вида $(0, a, b)$ или $(a, 0, b)$, то есть данное множество векторов не является линейным подпространством $\mathbb{R}^{3}$.

б)

Совокупность всех векторов, являющиеся линейными комбинациями конечного набора векторов являются _линейной оболочкой_. Т.е.  являются подпространcтвом!
